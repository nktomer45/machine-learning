#  SCATTER PLOT  MATPLOTLIB  

```python
from matplotlib import pyplot as plt
import pandas as pd

data=pd.read_csv('Social_Network_Ads.csv',nrows=100)
```

```python
data.head()
```

```python
data.tail()
```

```python
data.isnull().sum().sum()
```

```python
data.notnull().sum().sum()
```

```python
x=data["Age"]
y=data["EstimatedSalary"]
z=data["User ID"]
```

```python
plt.scatter(x,y) # scatter plot are used to plot point data on a horizontal and vertical axis
plt.title("Salary",fontsize=20)
plt.xlabel("age",fontsize=15)
plt.ylabel("EstimatedSalary",fontsize=15)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,color="r")
plt.title("Salary",fontsize=20)
plt.xlabel("age",fontsize=15)
plt.ylabel("EstimatedSalary",fontsize=15)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,color="r",marker="*",s=200,linewidths=3,edgecolor="g",verts="<")
plt.title("Salary",fontsize=20)
plt.xlabel("age",fontsize=15)
plt.ylabel("EstimatedSalary",fontsize=15)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,color="r",marker="*",s=200,linewidths=3,edgecolor="g")
plt.scatter(x,z,color="k",marker="*",s=200,linewidths=3,edgecolor="yellow")         
plt.title("Salary",fontsize=20)
plt.xlabel("age",fontsize=15)
plt.ylabel("EstimatedSalary",fontsize=15)
```
