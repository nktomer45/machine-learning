# pandas merging  function

```python
import pandas as pd
```

```python
data=pd.DataFrame({'id':[1,2,3,4],
                 'class':[5,6,7,8]})
```

```python
data
```

```python
df=pd.DataFrame({'id':[1,2,3,4],
    'Number':[1,2,34,3]})
```

```python
df
```

```python
pd.merge(data,df,on='id') #you can see "id data is command IN BOTH List"then both are merge
```

```python
pd.merge(df,data,on='id')
```

```python
data=pd.DataFrame({'id':[1,2,3,5],
                 'class':[5,6,7,8]})
df=pd.DataFrame({'id':[1,2,3,4],
    'Number':[1,2,34,3]})
pd.merge(df,data,on='id',how='inner')
```

```python
pd.merge(df,data,on='id',how ='outer')
```

```python
pd.merge(df,data,on='id',how ='right')
```

```python
pd.merge(df,data,on='id',how ='left')
```

```python
pd.merge(df,data,on='id',how ='outer',indicator=True)
```

```python
data=pd.DataFrame({'id':['w','e','u','t'],
                 'class':[5,6,7,8]})
df=pd.DataFrame({'id':[1,2,3,4],
    'Number':[1,2,34,3]})
pd.merge(df,data,left_index=True,right_index=True)
```

```python
data=pd.DataFrame({'i':[1,2,3,5],
                 'class':[5,6,7,8]})
df=pd.DataFrame({'id':[1,2,3,4],
    'Number':[1,2,34,3]})
```

```python
display(data,df)
```

```python
data.join(df)
```

```python
data.join(df,how='inner')
```

```python
data.join(df,how='outer')
```

# join () function

```python
data=pd.DataFrame({'id':[1,2,3,5],
                 'class':[5,6,7,8]})
df=pd.DataFrame({'i':[11,12,14,13],
    'Number':[1,2,34,3]})
display(data,df)
```

```python
data.join(df)
```

```python
data.join(df,how='inner')
```

# append 

```python
data=pd.DataFrame({'id':[1,2,3,5],
                 'class':[5,6,7,8]})
df=pd.DataFrame({'id':[11,12,14,13],
    'class':[1,2,34,3]})
display(data,df)
```

```python
data.append(df)
```

```python
data.append(df,ignore_index=True) # it used for indexing 
```

```python
data.append(df,ignore_index=True,sort=False)
```
