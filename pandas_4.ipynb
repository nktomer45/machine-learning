# pandas fillna


```python
import pandas as pd
```

```python
pt=pd.read_csv('50_Startups.csv')
```

```python
pt.head()
```

```python
pa=pd.read_csv('50_Startups.csv',na_values='0.00')
```

```python
pa
```

```python
pa.isnull().sum()
```

```python
pa.isnull().sum().sum()
```

```python
pa.fillna(0)
```

```python
pa.fillna(2)
pa.head()
```

```python
pa.fillna({'R&D Spend':'none','Marketing Spend':''}) # in this method used dictionary for fill null value in particular columns 
```

```python
pa.fillna(method='ffill') 
# it fill previse value at null place
```

```python
pa.fillna(method='bfill') # it revese of above method
pa.head()
```

```python
pa.fillna(method='pad') 
pa.head()
```

```python
pa.fillna(method='ffill',axis=0) 
```

```python
pa.fillna(method='ffill',axis=1) 
```

```python
pa.fillna(5,inplace=True) 
```
